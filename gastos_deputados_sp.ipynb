```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

filename = 'gastos-deputados-2018.csv'
df = pd.read_csv(filename)
df['sgpartido'].value_counts()
sum(df['vlrliquido'])
df_f = df.groupby(['nummes', 'sgpartido']).apply(lambda x:x['vlrliquido'].sum())
%matplotlib inline
print ('GASTOS DOS DEPUTADOS ESTADUAIS DE S√ÉO PAULO NO ANO DE 2018 POR PARTIDO')
fig, axes = plt.subplots(nrows=12, ncols=1)
mes = 'Janeiro Fevereiro Mar√ßo Abril Maio Junho Julho Agosto Setembro Outubro Novembro Dezembro'.split()
for x in range(12):
    df_f[x+1].plot.barh(ax = axes[x], figsize=(10,160), title = (f'Despesas dos deputados em {mes[x]}'))
```
